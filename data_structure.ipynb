q1.What are data structures, and why are they important
- Data structures organize and store data efficiently for easy access and modification. They are vital for program performance, code clarity, and algorithm design by providing optimized ways to handle data, making complex information manageable and operations quicker.

q2.Explain the difference between mutable and immutable data types with examples?
- Mutable data types like lists, dictionaries, sets can be changed after creation. Immutable data types like strings, tuples, numbers cannot be modified after being created. For example, you can append to a list, but changing a string requires creating a new one.

q3.What are the main differences between lists and tuples in Python?
- The main differences are mutability and syntax. Lists are mutable, allowing changes after creation, and use square brackets []. Tuples are immutable, meaning they cannot be changed after creation, and use parentheses (). Tuples are generally faster and safer for fixed collections.

q4.Describe how dictionaries store data
- Dictionaries store data as key-value pairs. Each unique key is associated with a specific value. This allows for efficient retrieval of values based on their corresponding keys, similar to how you'd look up a word in a dictionary.

q5.Why might you use a set instead of a list in Python?
- Use a set for unique elements and fast membership testing in keyword. Sets automatically handle duplicates, unlike lists. They are unordered, so don't use them if element order matters.

q6.What is a string in Python, and how is it different from a list?
- A string is an immutable sequence of characters used for text. A list is a mutable sequence of elements of any type. Strings are for text, lists for collections of items that can be changed.

q7.How do tuples ensure data integrity in Python?
- tuples ensure data integrity because they are immutable. Once created, the elements of a tuple cannot be changed. This prevents accidental modification of data, making tuples suitable for storing collections of items that should remain constant.

q8.What is a hash table, and how does it relate to dictionaries in Python?
- A hash table is a data structure storing key-value pairs for fast lookups. Python dictionaries are implemented using hash tables. They use a hash function to compute an index for efficient storage and retrieval of key-value pairs.

q9.Can lists contain different data types in Python?
- yes, lists can  contain different data types in Python.

q10.Explain why strings are immutable in Python?
- Strings are immutable in Python because this design choice allows for optimizations, like sharing string data in memory. It also makes strings reliable as dictionary keys and elements in sets, as their hash value remains constant.

q11. What advantages do dictionaries offer over lists for certain tasks?
- Dictionaries offer faster lookups and retrieval of values using keys, unlike lists which require iterating through elements. They are ideal for representing structured data with named fields and for quick checks for the presence of a key.

q12.Describe a scenario where using a tuple would be preferable over a list?
- A tuple is preferable when you have a collection of items that should not change, like coordinates x, y or RGB color values red, green, blue. Their immutability ensures data integrity in such fixed collections.

q13.How do sets handle duplicate values in Python?
- Sets inherently store only unique elements. When you add duplicate values to a set, they are automatically ignored, and only one instance of the value is kept.

q14.How does the “in” keyword work differently for lists and dictionaries
- For lists, the in keyword checks if an element's value exists anywhere in the list. For dictionaries, in checks if a key exists among the dictionary's keys.

q15.Can you modify the elements of a tuple? Explain why or why not?
- No, we cannot modify the elements of a tuple. This is because tuples are immutable data types in Python. Once a tuple is created, its contents are fixed and cannot be changed.

q16.What is a nested dictionary, and give an example of its use case?
- A nested dictionary is a dictionary where the values are themselves dictionaries. A common use case is representing hierarchical data, like a JSON structure or a complex configuration. For example, a dictionary of users where each user's value is another dictionary with their details (name, age, city).

q17.Describe the time complexity of accessing elements in a dictionary?
- Accessing elements in a dictionary by key has an average time complexity of O(1) (constant time). In the worst case, it can be O(n) (linear time) if there are many hash collisions, but this is rare with good hash functions

q18.In what situations are lists preferred over dictionaries?
- Lists are preferred when you need an ordered collection of items, can have duplicate elements, and access elements by their position (index). They are also suitable when the order of elements is important or when iterating through all elements is common.

q19.Why are dictionaries considered unordered, and how does that affect data retrieva?
- Dictionaries are considered unordered because they do not maintain the insertion order of elements (in Python versions before 3.7, they were truly unordered; in 3.7+ they maintain insertion order, but the concept of "unordered" still applies in a theoretical sense regarding key-based access). This means you cannot rely on retrieving items by their position or index. Data retrieval is done solely by using the keys.

q20.Explain the difference between a list and a dictionary in terms of data retrieval.?
- Data retrieval in lists is done by index (positional). Data retrieval in dictionaries is done by key (associative). This means you access elements in a list based on their order, while you access values in a dictionary based on their unique key.
